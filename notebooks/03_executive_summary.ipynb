### 4.2 Data Quality Issues

Multiple structural data quality issues were identified during the analysis:

#### Issue 1 — Composite Field Encoding
The `product_owner_reference` field encodes **two distinct pieces of information** in a single string, separated by a semicolon:
```
PRODUCT_CODE; DD/MM/YYYY
```
Similarly, the `product_description` field concatenates product name, supplier, country of origin, and batch number. These must be parsed via regex, creating fragility: any format inconsistency causes data loss.

**Recommendation:** Normalize these fields into separate database columns.

#### Issue 2 — Incomplete Product Categorization
The Product Details sheet lacked disease category mappings for a subset of products. Without categorization, these items cannot be attributed to HIV, Malaria, or TB budgets.

**Resolution applied:** AI-assisted categorization using product descriptions recovered most missing classifications. The residual \$116,241 (2.7% of total value) remains unclassified and should be reviewed manually.

#### Issue 3 — Missing Consumption Rates
Not all products have monthly consumption estimates in the product reference data. Where consumption data is absent, **stockout forecasting is impossible**, creating blind spots in supply planning.

**Recommendation:** Integrate actual distribution records to derive empirical consumption rates for all active products.

#### Issue 4 — Ambiguous Stock Events
The storage report contains both `Opening Stock on Hand` and `Closing Stock on Hand` event types per batch. Using the wrong event type leads to double-counting. This analysis consistently uses **closing stock** as the definitive position.

**Recommendation:** Clarify event type definitions in system documentation and expose only the relevant view to end users.

## Methodology Notes

**Data Sources:**
- *Storage Report:* Closing inventory by batch, including quantity, value, and expiry date (snapshot: May 28, 2026)
- *Product Details:* Reference data with monthly consumption estimates and disease sub-category

**Analysis Approach:**
- Stock value aggregated by disease category using closing stock events
- HIV adult products filtered by sub-category (`Adults`) and adult classification field
- Stockout dates calculated as: *current quantity ÷ daily consumption rate*, projected from snapshot date
- Expiry risk bucketed into 8 time horizons (expired → 730+ days)
- FEFO rotation applied: expired batches excluded from usable stock

**Tools:** Python (pandas, plotly), Jupyter Notebook, Claude AI (product categorization for missing mappings)

**Limitations:**
- Consumption rates are reference estimates, not verified against actual distribution records
- Forecasts assume no in-transit shipments or emergency resupply
- Products without consumption data are excluded from stockout projections
