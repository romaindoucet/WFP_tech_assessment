```python
import os
import numpy as np
import pandas as pd
```

```python
DATA_FILE_PATH = "../data/Technical Assignment_Data_dev.xlsx"
SHEETS = ["Storage Report", "Product Details"]
```

```python
df_storage_raw = pd.read_excel(DATA_FILE_PATH, sheet_name=SHEETS[0])
df_storage_raw.head()
```

```python
df_product_details_raw = pd.read_excel(DATA_FILE_PATH, sheet_name=SHEETS[1])
df_product_details_raw.head()
```

```python
df = (
    df_product_details_raw
    .pipe(lambda d: d[d.product_code.str.startswith("SMSUCOND")])
    .sort_values("product_code")
    #.query("product_code == 'SMSUCOND1'")
)
df.head()
```

```python
df_product_details = df_product_details_raw.rename(columns={
    "Description": "product_description",
    "Category": "product_category",
    "Sub-Category": "product_sub_category",
    "Acronym": "product_acronym",
    "Monthly Consumption (# units)": "monthly_consumption_in_units_per_month",
})
df_product_details.head()
```

```python
df_storage_raw.head()
```

```python
df_storage = df_storage_raw.rename(columns={
    "event label": "event_label",
    "date value": "event_date",
    "event line IU quantity": "qty_inventory_unit",
    "inventory unit (IU)": "iu",
    "event line HU quantity": "qty_handling_unit",
    "handling unit (HU)": "handling_unit",
    "event line weight (mT)": "weight_mt",
    "event line volume (m3)": "volume_m3",
    "event line value (USD)": "value_usd",
    "product category": "product_category_storage",
    "product owner reference": "product_owner_ref_raw",
    "product description": "product_description_storage",
})

```

```python
df_storage_raw.head(2)
```

```python
parsed = df_storage["product_owner_ref_raw"].str.split(";", expand=True)
df_storage["product_code"] = parsed[0].str.strip()
df_storage["expiry_date"] = pd.to_datetime(parsed[1].str.strip(), format="%d/%m/%Y", errors="coerce")

df_storage[["product_owner_ref_raw", "product_code", "expiry_date"]].head(10)
```

```python
df_storage
```

```python
df_storage = df_storage[
    [
        "product_code",
        "expiry_date",
        "qty_inventory_unit",
        "value_usd",
        "event_date",
        "event_label",
        "iu",
        "qty_handling_unit",
        "handling_unit",
        "weight_mt",
        "volume_m3",
        "product_description_storage",
        "product_category_storage",
        "product_owner_ref_raw",
    ]
].astype({
    "qty_inventory_unit": "Int64",
    "qty_handling_unit": "Int64",
    "value_usd": "Int64",
})

df_product_details = df_product_details[
    [
        "product_code",
        "product_category",
        "product_sub_category",
        "product_acronym",
        "product_description",
        "monthly_consumption_in_units_per_month",
    ]
]
```

```python
# % valeurs manquantes par colonne
print("=== df_storage ===")
print((df_storage.isna().mean() * 100).round(1).to_string())
print()
print("=== df_product_details ===")
print((df_product_details.isna().mean() * 100).round(1).to_string())
```

```python
# Références dans df_storage absentes de df_product_details
known_codes = set(df_product_details["product_code"])
mask_unknown = ~df_storage["product_code"].isin(known_codes)

unknown = df_storage[mask_unknown]
total = df_storage

summary = pd.DataFrame({
    "n_rows": [mask_unknown.sum(), len(df_storage)],
    "n_unique_codes": [unknown["product_code"].nunique(), total["product_code"].nunique()],
    "qty_inventory_unit": [unknown["qty_inventory_unit"].sum(), total["qty_inventory_unit"].sum()],
    "value_usd": [unknown["value_usd"].sum(), total["value_usd"].sum()],
}, index=["not in product_details", "total"])

summary["pct_qty"] = (summary["qty_inventory_unit"] / summary.loc["total", "qty_inventory_unit"] * 100).round(1)
summary["pct_value"] = (summary["value_usd"] / summary.loc["total", "value_usd"] * 100).round(1)

summary
```

```python
df = df_storage.merge(df_product_details, on="product_code", how="left")
df.shape
```
